## Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import load_model
```

## Initailize the dataframe

```python
df = pd.read_csv('fake_reg.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

## EDA

```python
sns.pairplot(df)
plt.show()
```

```python
sns.heatmap(data=df.corr(), annot=True, cmap='Greens', cbar=False)
plt.show()
```

## Model Training - Artificial Neuron Network

```python
np.random.seed(42)
```

```python
X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

#### Scaling down the data (normalize the data)

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.max(), X_train.min()
```

#### Creating a Neural Network

```python
# Method 1
model = Sequential([Dense(units=4, activation='relu'),
                   Dense(units=2, activation='relu'),
                   Dense(units=1)])
```

```python
# Method 2
model = Sequential()

model.add(Dense(units=4, activation='relu'))
model.add(Dense(units=4, activation='relu'))
model.add(Dense(units=4, activation='relu'))

model.add(Dense(units=1))
```

```python
model.compile(optimizer='rmsprop', loss='mse')
```

```python
model.fit(x=X_train, y=y_train, epochs=250, verbose=True)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.head()
```

```python
sns.lineplot(loss_df.loss)
plt.show()
```

```python
model.evaluate(X_test, y_test, verbose=False)
```

```python
model.evaluate(X_train, y_train, verbose=False)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = pd.Series(y_pred.reshape(300,))
```

```python
pred_df = pd.DataFrame(y_test, columns=['Test True Y'])
pred_df.head()
```

```python
pred_df = pd.concat([pred_df, y_pred], axis=1)
pred_df.columns = ['True Y Values', 'Model Y Predictions']
```

```python
pred_df.head()
```

#### Evaluation

```python
sns.scatterplot(data=pred_df, x='True Y Values', y='Model Y Predictions')
plt.show()
```

```python
mean_absolute_error(pred_df['True Y Values'], pred_df['Model Y Predictions'])
```

```python
mean_squared_error(pred_df['True Y Values'], pred_df['Model Y Predictions'])
```

```python
np.sqrt(mean_squared_error(pred_df['True Y Values'], pred_df['Model Y Predictions']))
```
